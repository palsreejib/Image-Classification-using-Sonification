# Spectrogram Representation


## Generate and Display a Linear-Scale Spectrogram from Audio

This code snippet loads an audio waveform and generates its **spectrogram** using the **Short-Time Fourier Transform (STFT)**. It then visualizes the spectrogram using a **linear frequency scale**, which preserves the original frequency spacing (unlike mel or log scales).

### Key Steps:
- Loads an audio file using `librosa`
- Applies STFT to extract the frequency domain representation over time
- Converts the amplitude spectrogram to a decibel (dB) scale for better visual contrast
- Plots the spectrogram using `matplotlib` with linear frequency resolution

This visualization is useful for analyzing raw frequency content, especially for lower-frequency signals like machinery, engine sounds, or speech formants.


```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load audio file
y, sr = librosa.load(r'D:\Documents\Iquisitionis\103\car.wav') 

# Perform STFT
stft = np.abs(librosa.stft(y))

# Convert to dB scale
stft_db = librosa.amplitude_to_db(stft, ref=np.max)

# Create the spectrogram plot with linear frequency scale
plt.figure(figsize=(12, 6))
librosa.display.specshow(stft_db, sr=sr, x_axis='time', y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram (Linear Frequency Scale)')
plt.ylabel('Frequency (Hz)')
plt.ylim(0, sr/2)  # Show frequencies up to Nyquist frequency
plt.show()
```
